```python
#assign tag columns
for i in gift:
  df.loc[df['title'].str.contains(gift_dict[i]), i] = 1
df
```

```python
#assign category columns
for i in gift:
  df.loc[df['title'].str.contains(gift_dict[i]), 'title_category'] = str(i)
df
```

```python
#add 'sum' colum for modifying mix category
col_list = ['spam','tuna','bath','oil','ginseng','coffee','nuts','dried_seafood','meat','grains','gift_money','snacks','fruits','perfume','cosmetics','spicy','nutri','tools','hygiene','man','babies']
df["sum"] = df[col_list].sum(axis=1)
df
```

```python
#modifying mix category with sum(tags)
df.loc[(df['sum'] > 1) | (df['mix'] == 1), 'title_category'] = str('mix')
```
